```python
import numpy as np
from scipy import stats
```

```python
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=100)  
group2 = np.random.normal(loc=53, scale=10, size=100) 
```

```python
t_stat, p_value = stats.ttest_ind(group1, group2)
```

```python
print("Group 1 mean:", np.mean(group1))
print("Group 2 mean:", np.mean(group2))
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")
```

```python
if p_value < 0.05:
    print("✅ The difference in means is statistically significant (p < 0.05).")
else:
    print("❌ No significant difference in means (p >= 0.05).")
```
