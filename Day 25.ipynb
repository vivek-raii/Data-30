```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('Titanic-Dataset.csv')
num_cols = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived', 'Pclass']
# Quick look
df.info()
df.head()
```

```python
corr = df[num_cols].corr()
print(corr)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap (Numerical Features)', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title('Age vs Fare by Survival')

plt.subplot(1,2,2)
sns.scatterplot(data=df, x='SibSp', y='Parch', hue='Survived', palette='Set2')
plt.title('SibSp vs Parch by Survival')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[num_cols], hue='Survived', palette='husl', diag_kind='kde')
plt.suptitle('Pairwise Relationships of Numerical Features', y=1.02)
plt.show()
```
