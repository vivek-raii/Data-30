```python
import matplotlib.pyplot as plt
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales': [12000, 13500, 15000, 17000, 18500, 21000,
              19500, 22000, 23000, 25000, 27000, 30000]
}
```

```python
plt.figure(figsize=(8,5))
plt.plot(data["Month"], data["Sales"], marker='o', linestyle='-', color='b')
plt.title("Monthly Sales Growth in 2025", fontsize=14, fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Sales (in )", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
